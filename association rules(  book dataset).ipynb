# apriori algorithm

```python
from mlxtend.frequent_patterns import apriori
frequent_itemset=apriori(books,min_support=0.1,use_colnames=True)
frequent_itemset
```

# association rules

```python
from mlxtend.frequent_patterns import association_rules
Rules=association_rules(frequent_itemset,metric="lift",min_threshold=0.7)
Rules
```

```python
Rules[Rules.lift>1]
```

```python
Rules=association_rules(frequent_itemset,metric="confidence",min_threshold=0.6)
Rules
```
