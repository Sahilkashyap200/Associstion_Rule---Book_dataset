# apriori algorithem

```python
from mlxtend.frequent_patterns import apriori
frequent_itemset=apriori(movies_,min_support=0.2,use_colnames=True)
frequent_itemset
```

# Association Rules

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemset,metric="confidence",min_threshold=0.6)
rules
```

```python
rules[rules.lift>1]
```

```python
rules.sort_values('confidence',ascending=True)
```
